```python
# importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#importing the dataset
train_set = pd.read_csv("train.csv")
test_sets = pd.read_csv("test.csv")
```

```python
test_sets
```

```python
train_set  = train_set .drop(['PassengerId','Name','Cabin','Ticket',], axis=1)
```

```python
train_set  
```

```python
test_set  = test_sets .drop(['PassengerId','Name','Cabin','Ticket',], axis=1)
test_set
```

```python
#  taking care of categorical data
train_set.isnull().sum()
```

```python
#from sklearn.preprocessing import Imputer
#imputer = Imputer(missing_values='NAN',strategy = "mean", axis =0)
#imputer = imputer.fit(train_set[])

train_set['Age'] = train_set['Age'].fillna(train_set['Age'].mean())
train_set['Embarked'] = train_set['Embarked'].fillna("D")
```

```python
train_set.isnull().sum()
```

```python
#  taking care of categorical data
test_set.isnull().sum()
```

```python
test_set['Age'] = train_set['Age'].fillna(train_set['Age'].mean())
test_set['Fare'] = train_set['Fare'].fillna(train_set['Fare'].mean())
```

```python
#  taking care of categorical data
test_set.isnull().sum()
```

```python
train_set
```

```python
test_set
```

```python
X = train_set.iloc[:,1:].values
Y= train_set.iloc[:,0].values

test = test_set.iloc[:, 0:].values
```

```python
X
```

```python
# Encoding categorical data
# Encoding the Independent Variable
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
labelencoder_X = LabelEncoder()
X[:,1] = labelencoder_X.fit_transform(X[:,1])
X[:,6] = labelencoder_X.fit_transform(X[:,6])
onehotencoder1 = OneHotEncoder(categorical_features = [6])
X = onehotencoder1.fit_transform(X).toarray()


#encoding the categorical features in the test set
labelencoder_test = LabelEncoder()
test[:, 1] = labelencoder_test.fit_transform(test[:, 1])
test[:, 6] = labelencoder_test.fit_transform(test[:, 6])

onehotencoder2 = OneHotEncoder(categorical_features = [6])
test = onehotencoder2.fit_transform(test).toarray()
```

```python
#avoiding dummy variable trap
X = X[:, 1:]

X
```

```python
#splitting the dataset into train and test set
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.1, random_state= 0)
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
X_train = sc_x.fit_transform(X_train)
X_test = sc_x.transform(X_test)

test = sc_x.fit_transform(test)
```

```python
test
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

regressor = LogisticRegression()
regressor.fit(X_train, Y_train)
Y_pred = regressor.predict(X_test)
acc_regressor = round(accuracy_score(Y_pred, Y_test) * 100, 2)
print(acc_regressor)
```
